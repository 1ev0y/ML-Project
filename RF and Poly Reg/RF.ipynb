```python
import Helper
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
data_set = Helper.load_data()
data_set = Helper.proc_data(data_set, del_columns = ['DATE'])
data_set.head()
import pandas as pd
```

```python
x_train, x_test, y_train, y_test = Helper.split_data(data_set, 'GDP')
```

```python
from sklearn.ensemble import RandomForestRegressor

def perform_RF(x_train, x_test, y_train, y_test):
    model = RandomForestRegressor()

    parameters = {
        'n_estimators' : [100, 200, 300],
        'max_depth' : [None, 10, 30],
        'criterion' : ['squared_error', 'friedman_mse'],
        'min_samples_split' : [2, 10, 20],
        'min_samples_leaf' : [1, 5, 10],
        'max_features' : ['sqrt', 'log2', 0.5]
    }

    Helper.Grid_Search(model, x_train, y_train, x_test, y_test, parameters)
```

```python
perform_RF(x_train, x_test, y_train, y_test)
```

Trying to improve even more. Checking baseline Performance.

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

fine_tune_model = RandomForestRegressor(criterion= 'friedman_mse',max_depth= None, max_features= 0.5, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 300)
fine_tune_model.fit(x_train, y_train)

y_pred = fine_tune_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print(f"MSE -> {mse}")
```

Increasing number of estimators.

```python
fine_tune_model = RandomForestRegressor(criterion= 'friedman_mse',max_depth= None, max_features= 0.5, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 1000)
fine_tune_model.fit(x_train, y_train)

y_pred = fine_tune_model.predict(x_test)

mse = mean_squared_error(y_test, y_pred)
print(f"MSE -> {mse}")
```

```python
model = RandomForestRegressor(criterion= 'squared_error',max_depth= None, max_features= 0.5, min_samples_leaf= 1, min_samples_split= 2, n_estimators= 1000)

train_data_set = Helper.load_data('2_Y_Lag','../Datasets/Lagged/')
test_data_set = Helper.load_data('Lagged_5Y','../Datasets/Lagged/')
train_data_set = Helper.proc_data(train_data_set)
train_data_set['DATE'] = pd.to_datetime(train_data_set['DATE'], format="%d-%m-%Y")
test_data_set['DATE'] = pd.to_datetime(test_data_set['DATE'], format="%d-%m-%Y")

test_data_set = test_data_set[test_data_set['DATE'] <= '2024-03-02']
test_data_set = test_data_set[test_data_set['DATE'] >= '2022-03-01']

train_data_set = train_data_set[train_data_set['DATE'] < '2022-03-01']

target_var = 'GDP'
feature_vars = train_data_set.columns.to_list()
feature_vars.remove(target_var)
feature_vars.remove('DATE')

x_train = train_data_set[feature_vars].to_numpy()
y_train = train_data_set[target_var].to_numpy()
x_test = test_data_set[feature_vars].to_numpy()
   
model.fit(x_train, y_train)

y_pred = model.predict(x_test)

# print(f"MSE -> {mean_squared_error(y_test, y_pred)}")

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dates = pd.date_range(start='2024-03-01', periods=len(y_pred), freq='D')

plt.figure(figsize=(10, 6))
plt.plot(dates, y_pred, linestyle='-', color='b', label='Predicted GDP')
# plt.plot(dates, y_test, linestyle='-', color='r', label='Actual GDP')
plt.title('Predicted GDP Over Time', fontsize=14)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Predicted GDP', fontsize=12)
plt.xticks(rotation=45)  
plt.grid(True)

# Show plot
plt.tight_layout()
plt.legend()
plt.show()
```
